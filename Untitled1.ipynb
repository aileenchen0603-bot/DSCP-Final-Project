```python

# Step 1:Import the database

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import matplotlib.font_manager as fm
import warnings

try:
    from pandas.core.api import SettingWithCopyWarning
    warnings.filterwarnings('ignore', category=SettingWithCopyWarning)
except ImportError:

    print("Warning: Could not import SettingWithCopyWarning. Analysis proceeds.")

print("Initializing environment and loading libraries...")


try:
    fm.fontManager.clear()
    print("Matplotlib font cache cleared successfully.")
except AttributeError:
    pass


sns.set_theme(style="whitegrid")
plt.rcParams['axes.unicode_minus'] = False

# Import JHU GitHub
confirmed_url = "https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_confirmed_global.csv"
deaths_url = "https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_deaths_global.csv"

# Read CSV file
try:
    df_confirmed = pd.read_csv(confirmed_url)
    df_deaths = pd.read_csv(deaths_url)
    print("Data loaded successfully from JHU GitHub.")
except Exception as e:
    print(f"Data loading failed: {e}")


# Step 2: Data cleaning and format conversion

def clean_and_melt(df, value_name):
    """Converts wide-format time series data to long format and aggregates to Country level."""
    df = df.drop(columns=['Lat', 'Long'])
    df_melted = df.melt(
        id_vars=['Province/State', 'Country/Region'],
        var_name='Date',
        value_name=value_name
    )
    df_melted['Date'] = pd.to_datetime(df_melted['Date'])
    # Aggregate data to the Country level
    df_grouped = df_melted.groupby(['Date', 'Country/Region']).sum().reset_index()
    return df_grouped


df_confirmed_clean = clean_and_melt(df_confirmed, 'Confirmed')
df_deaths_clean = clean_and_melt(df_deaths, 'Deaths')


df_full = pd.merge(df_confirmed_clean, df_deaths_clean, on=['Date', 'Country/Region'])

df_full['CFR'] = (df_full['Deaths'] / df_full['Confirmed']) * 100

print(f"Data cleaning complete. Total records: {len(df_full)}")


# step 3: Global Cumulative Trend

df_global = df_full.groupby('Date')[['Confirmed', 'Deaths']].sum().reset_index()


plt.figure(figsize=(15, 6))

# Global Confirmed Cases
plt.subplot(1, 2, 1)
sns.lineplot(x='Date', y='Confirmed', data=df_global)
plt.title('Global Cumulative Confirmed Cases (Log Scale)', fontsize=14)
plt.xlabel('Date')
plt.ylabel('Cumulative Cases (Log)')
plt.yscale('log')
plt.grid(True)

# Global Death Cases
plt.subplot(1, 2, 2)
sns.lineplot(x='Date', y='Deaths', data=df_global, color='red')
plt.title('Global Cumulative Deaths (Log Scale)', fontsize=14)
plt.xlabel('Date')
plt.ylabel('Cumulative Deaths (Log)')
plt.yscale('log')
plt.grid(True)

plt.tight_layout()
plt.savefig('Chart_3_Global_Cumulative_Trend.png', dpi=300)
plt.show()

# Step 4: Top 10 Countries Comparison


latest_date = df_full['Date'].max()
df_latest = df_full[df_full['Date'] == latest_date].sort_values(by='Confirmed', ascending=False)
top_countries = df_latest.head(10)['Country/Region'].tolist()
df_top_countries = df_full[df_full['Country/Region'].isin(top_countries)]

plt.figure(figsize=(14, 7))
sns.lineplot(
    x='Date',
    y='Confirmed',
    hue='Country/Region',
    data=df_top_countries,
    legend='full'
)
plt.title(f'Cumulative Confirmed Cases: Top 10 Countries (As of {latest_date.strftime("%Y-%m-%d")})', fontsize=14)
plt.xlabel('Date')
plt.ylabel('Cumulative Confirmed Cases')
plt.legend(title='Country/Region', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True)
plt.tight_layout()
plt.savefig('Chart_4_Top10_Cumulative_Comparison.png', dpi=300)
plt.show()

# Step 5: CFR analysis


df_latest_top10 = df_latest[df_latest['Country/Region'].isin(top_countries)].copy()
df_latest_top10['Total_CFR'] = (df_latest_top10['Deaths'] / df_latest_top10['Confirmed']) * 100


plt.figure(figsize=(12, 6))
sns.barplot(
    x='Total_CFR',
    y='Country/Region',
    data=df_latest_top10.sort_values(by='Total_CFR', ascending=False),
    palette='viridis'
)
plt.title(f'Case Fatality Ratio (CFR) Comparison: Top Confirmed Countries', fontsize=14)
plt.xlabel('Cumulative CFR (%)')
plt.ylabel('Country/Region')
plt.grid(axis='x')
plt.savefig('Chart_5_CFR_Top_Countries.png', dpi=300)
plt.show()



# Step 6: Daily New Cases analysis


df_global['Daily_New_Cases'] = df_global['Confirmed'].diff().fillna(df_global['Confirmed'])


plt.figure(figsize=(12, 6))
sns.lineplot(x='Date', y='Daily_New_Cases', data=df_global, color='darkorange')
plt.title('Global Daily New Confirmed Cases Trend', fontsize=14)
plt.xlabel('Date')
plt.ylabel('Daily New Cases')
plt.grid(True)
plt.savefig('Chart_6_Global_Daily_New_Cases.png', dpi=300)
plt.show()


target_countries = ['Taiwan*', 'Japan', 'US', 'Germany']

plt.figure(figsize=(15, 8))

for country in target_countries:
    df_country = df_full[df_full['Country/Region'] == country].copy()

    if df_country.empty:
        print(f"Warning: Data for '{country}' not found. Skipping.")
        continue

    df_country['Daily_New_Cases'] = df_country['Confirmed'].diff().fillna(df_country['Confirmed'])

    sns.lineplot(
        x='Date',
        y='Daily_New_Cases',
        data=df_country,
        label=country
    )

plt.title('Comparison of Daily New Confirmed Cases', fontsize=14)
plt.xlabel('Date')
plt.ylabel('Daily New Cases')
plt.legend(title='Country/Region')
plt.grid(True)
plt.tight_layout()
plt.savefig('Chart_6_Selected_Daily_New_Comparison.png', dpi=300)
plt.show()


# Save the data

print("\n--- Saving Data Files ---")


df_full.to_csv('COVID_Data_Cleaned_Full.csv', index=False, encoding='utf-8')
print("Saved: COVID_Data_Cleaned_Full.csv")

df_global.to_csv('COVID_Data_Global_Daily_Summary.csv', index=False, encoding='utf-8')
print("Saved: COVID_Data_Global_Daily_Summary.csv")

```
